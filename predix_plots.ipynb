```python
import pandas as pd
from bokeh.io import push_notebook, show, output_notebook
from bokeh.charts import Scatter
```

## Working on Wind Turbine 2

```python
wt2 = pd.read_csv("WT002.csv")
```

```python
data = dict(active_power = wt2['WT002_ACTIVE_POWER'], turbine_status=wt2['WT002_TURBINE_STATUS'])
p = Scatter(
  data, y='active_power', x='turbine_status', ylabel='Active Power',
  xlabel='Turbine Status', title='Active Power vs Turbine Status', color='green'
)

output_notebook()
show(p)
```

```python
data = dict(active_power = wt2['WT002_ACTIVE_POWER'], reactive_power=wt2['WT002_REACTIVE_POWER'])
p = Scatter(
  data, y='active_power', x='reactive_power', ylabel='Active Power',
  xlabel='Reactive Power', title='Active Power vs Reactive Power', color='blue'
)

output_notebook()
show(p)
```
