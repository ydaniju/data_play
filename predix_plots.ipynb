```python
import pandas as pd
from bokeh.io import push_notebook, show, output_notebook
from bokeh.charts import Scatter
```

```python
data = dict(active_power = wt02['WT002_ACTIVE_POWER'], turbine_status=wt02['WT002_TURBINE_STATUS'])
p = Scatter(
  data, y='active_power', x='turbine_status', ylabel='Active Power',
  xlabel='Turbine Status', title='Active Power vs Turbine Status for turbine 2', color='green'
)

output_notebook()
show(p)
```
